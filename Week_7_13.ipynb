```python
import heapq

N, M = map(int, input().split())

graph = [[] for _ in range(N)]

for _ in range(M):
    u, v, w = map(int, input().split())
    graph[u].append((w, v))
    graph[v].append((w, u))

visited = [False] * N
min_heap = [(0, 0)]
total_cost = 0

while min_heap:
    cost, node = heapq.heappop(min_heap)

    if visited[node]:
        continue

    visited[node] = True
    total_cost += cost

    for edge_cost, neighbor in graph[node]:
        if not visited[neighbor]:
            heapq.heappush(min_heap, (edge_cost, neighbor))

print("Minimum Cost of MST =", total_cost)
```
