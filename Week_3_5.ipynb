```python
T = int(input())

for _ in range(T):
    N, S = map(int, input().split())
    arr = list(map(int, input().split()))
    mid = N // 2
    left = arr[:mid]
    right = arr[mid:]

    left_sums = []
    right_sums = []
    for mask in range(1 << len(left)):
        total = 0
        for i in range(len(left)):
            if mask & (1 << i):
                total += left[i]
        left_sums.append(total)

    for mask in range(1 << len(right)):
        total = 0
        for i in range(len(right)):
            if mask & (1 << i):
                total += right[i]
        right_sums.append(total)

    right_sums = set(right_sums)
    answer = "NO"
    for x in left_sums:
        if S - x in right_sums:
            answer = "YES"
            break

    print(answer)
```
