```python
def create_trie():
    return {}
def insert(trie, word):
    node = trie
    for char in word:
        if char not in node:
            node[char] = {}
        node = node[char]
    node['#'] = True
def search(trie, word):
    node = trie
    for char in word:
        if char not in node:
            return False
        node = node[char]
    return '#' in node
def starts_with(trie, prefix):
    node = trie
    for char in prefix:
        if char not in node:
            return False
        node = node[char]
    return True
contacts = ["Anil", "Anitha", "Anirudh", "Bala", "Balaji"]

trie = create_trie()
for name in contacts:
    insert(trie, name)

# Operations
print('Contact "Anil" found →', search(trie, "Anil"))
print('Contact "Anand" found →', search(trie, "Anand"))
print('Prefix "Ani" exists →', starts_with(trie, "Ani"))
print('Prefix "Bal" exists →', starts_with(trie, "Bal"))
```
