```python
n, m, e = map(int, input().split())

graph = [[] for _ in range(n + 1)]

for _ in range(e):
    u, v = map(int, input().split())
    graph[u].append(v)

matchR = [-1] * (m + 1)

def dfs(u, visited):
    for v in graph[u]:
        if not visited[v]:
            visited[v] = True
            if matchR[v] == -1 or dfs(matchR[v], visited):
                matchR[v] = u
                return True
    return False

max_matching = 0

for u in range(1, n + 1):
    visited = [False] * (m + 1)
    if dfs(u, visited):
        max_matching += 1

print("Maximum Matching Size =", max_matching)

print("Matching:")
for v in range(1, m + 1):
    if matchR[v] != -1:
        print(matchR[v], v)
```
