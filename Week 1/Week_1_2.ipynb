```python
def max_crossing_sum(arr, left, mid, right):
    left_sum = float('-inf')
    s = 0
    for i in range(mid, left - 1, -1):
        s += arr[i]
        left_sum = max(left_sum, s)
    # Include elements on right of mid
    right_sum = float('-inf')
    s = 0
    for i in range(mid + 1, right + 1):
        s += arr[i]
        right_sum = max(right_sum, s)

    return left_sum + right_sum


def max_subarray_sum(arr, left, right):
    if left == right:
        return arr[left]

    mid = (left + right) // 2

    return max(
        max_subarray_sum(arr, left, mid),          # left subarray
        max_subarray_sum(arr, mid + 1, right),     # right subarray
        max_crossing_sum(arr, left, mid, right)    # crossing subarray
    )

t = int(input().strip())
for i in range(t):
    n = int(input().strip())
    arr = list(map(int, input().split()))
    result = max_subarray_sum(arr, 0, n - 1)
    print(result)
```
